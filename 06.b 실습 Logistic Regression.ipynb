# 분류 모델 실습

## pima-indians-diabetes.csv 파일을 읽어서, 당뇨병을 분류하는 모델을 만드시오.

컬럼 정보 :

Preg=no. of pregnancy

Plas=Plasma

Pres=blood pressure

skin=skin thickness

test=insulin test

mass=body mass

pedi=diabetes pedigree function

age=age

class=target(diabetes of not, 1:diabetic, 0:not diabetic)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('../data/pima-indians-diabetes.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.loc[ : , 'Plas':'mass' ] = df.loc[ : , 'Plas':'mass' ].replace( 0 , np.nan )
```

```python
df
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
y = df['class']
```

```python
X = df.drop('class', axis = 1)
```

```python
X.head(3)
```

```python
y.head(2)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
scaler_X = StandardScaler()
```

```python
X_scaled = scaler_X.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2,random_state= 11)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
(47+10)/cm.sum()
```
